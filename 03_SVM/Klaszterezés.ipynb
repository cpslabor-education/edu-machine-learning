## Kmeans

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=1000, centers=8, random_state=0)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.score(X, y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_.astype(float))
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='+')
```

## 3D adat

```python
X, y = make_blobs(n_samples=1000, centers=8, n_features=3, random_state=42)
```

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot(projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], c=y)
```

```python
kmeans = KMeans(n_clusters=8, random_state=0).fit(X)
```

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot(projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], c=kmeans.labels_.astype(float))
ax.scatter(kmeans.cluster_centers_[:, 0], 
            kmeans.cluster_centers_[:, 1], 
            kmeans.cluster_centers_[:, 2], 
            c='b',marker='+')
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
X, y = make_blobs(n_samples=1000, centers=8, n_features=3, random_state=42)
X = StandardScaler().fit_transform(X)
```

```python
db = DBSCAN(eps=0.3, min_samples=10).fit(X)
```

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot(projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], c=db.labels_.astype(float))
```

```python
import numpy as np
```

```python
for i,e in enumerate(np.arange(0.1, 1.0, 0.1)):
    fig = plt.figure(figsize=(3,3))
    db = DBSCAN(eps=e, min_samples=10).fit(X)
    ax = fig.add_subplot(projection='3d')
    ax.scatter(X[:,0], X[:,1], X[:,2], c=db.labels_.astype(float))
```

## Spektr√°lis klaszterez√©s

```python
from sklearn.cluster import SpectralClustering
```

```python
X, y = make_blobs(n_samples=1000, centers=8, n_features=3, random_state=42)
```

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot(projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], c=y)
```

```python
sp = SpectralClustering(n_clusters=8, 
                        assign_labels='discretize', 
                        random_state=0).fit(X)
```

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot(projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], c=sp.labels_.astype(float))
```

```python
sp = SpectralClustering(n_clusters=3, 
                        assign_labels='discretize', 
                        random_state=0).fit(X)
fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot(projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], c=sp.labels_.astype(float))
```
